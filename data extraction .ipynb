```python
ticker = 'AAPL'
multiplier  ='1'
timespan='day'
start='2010-10-14'
end='2022-10-14'
limit='50000'
API='...'
url=f'https://api.polygon.io/v2/aggs/ticker/{ticker}/range/{multiplier}/{timespan}/{start}/{end}?&limit={limit}&apiKey={API}'
url
```

```python
# api call 
```

```python
import requests 
import pandas as pd
import glob 
import os 

data= requests.get(url).json()
data

```

```python
df = pd.DataFrame(data['results'])
df['date'] = pd.to_datetime(df['t'],unit='ms')#unit='ms'
df.rename(columns={0:"volume",
                   1:"weighted volume",
                   2:"open",
                   3:"close",
                   4:"high",
                   5:"low",
                   6:"timestamp",
                   7:"transactions",
                   8:"date"},
         inplace=True)

df
```

```python
df.to_csv(ticker+start+('.csv'),header=None)
```
